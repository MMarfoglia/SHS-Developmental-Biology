## Morphogens 

###  1. Definition of a morphogen

Morphogens are described as secreted signaling molecules that diffuse from local sources to form concentration gradients, which in turn specifiy multiple cell fates depending. This concentration gradient induces in a different manner genetic responses allowing to spatially modify cell fate. Morphogens are often associated with lipids to improve their diffusion ability.

### 2. Process in which morphogens are involved

Morphogens are involved in a lot of physiological processes. During development, they define the major axes of the embryo (antero-posterior, dorso-ventral axis). Bicoid is one of the most studied morphogens, as it allows the establishement of the first embryo axis: the antero-posterior axis.

Scientists found bicoid mutants where the antero-posterior segmentation failed by performing a siRNA screen. The inhibition of bicoid results into a miss-segmentation of the embryo, which has 2 posterior parts. 

PHOTO EMBRYON WT + MUTANT BICOID

The Drosophila Melanogaster anterior-posterior axis is established during oogenesis by the localization of bicoid mRNA at the posterior pole of the oocyte. The egg is composed of nurse cells and the oocyte. At this stage, the oocyte is not able to express its own genes. Nurse cells thus provide to the oocyte the required nutritional resources but also key genes required for early segmentation of the oocyte, such as the establishement of the anterior-posterior axis. Bicoid is one of these genes. It is transcribed from the nurse cells' genome and the resulting mRNA is transported into the oocyte. 

The establishement of the anterior-posterior axis relies on the position of the oocyte nuclei. The nuclei position determined the dorsal part of the embryo, thus determining the dorso-ventral axis. Once the dorsal part of the embryo is established, the anterio-posterior axis can be established. 

Now that the antero-posterior axis is determined, this will activate a microtubule-dependent active transport of the bicoid mRNA towards the anterior part of the embryo.

Bicoid mRNAs form an equatorial plate at the anterior part of the embryo. The first established model was that this equatorial plate corresponds to the place where the bicoid mRNA will be translated into proteins. The proteins will then diffuse forming the morphogen gradient. Cells are potents to receive the signal corresponding to the bicoid gradient. Potency means that these cells are able to sense bicoid concentration but also to respond to it in a manner that is concentration dependent. This will create the first segmentation of the embryo along the antero-posterior axis.

However, when scientists tried to go deeper into the understanding of this gradient, they realized that it was not a sufficient way for segmenting the embryo. Indeed, if bicoid was passively diffusing from the anterior part of the embryo toward the posterior part, the gradient would not be fast enough to segment in time the embryo whose length is 2mm.

## III. Morphogen interpretation: a context-dependent process

From what we have seen so far we know that morphogens acts on receiving cells by inducing a transcriptionnal response. however only a limited number of inductive signals are iteratively used during development to coordinate tissue patterning.This implies that the same morphogen can have different effects on receiving cells depending on the developmental context. 

To understand how this is possible we need to have a look to the mechanism of action by which morphogen specify cells.

Thus, we easily understand that the transcriptional response of cells not only depends on the availability of the morphogen molecules, which could be referred as **signaling level and gradient shape**, but also on the ability of the cells to **sense** and **interpret** these molecules, which could be referred as **signal interpretation**. Therefore the action of morphogen on tissue is **highly context dependent**.

In this section we will explore the different parameters that can influence the way the cells can respond to morphogen.

### Influence of signal level and morphogen distribution

By definition, the response of receiving cell to morphogen depends on their distance to the source. It suggests that the morphogen concentration determines the downstream transcriptional program of receiving cells in a dose dependant manner. It means changes in extracellular morphogen concentration is transduced as differences in the activity of relevant transcription factors. For example the switch in gene expression in Xenopus cells caused by a threefold difference in activin concentration (morphogen) could be reproduced without morphogen by inducing a comparable change in the level of the downstream transcription factor Smad2. The changes in transcriptional program provides the positional information to the cell leading to tissue patterning.

Generally a single type of morphogen binding receptor is sufficient to specify concentration-dependent response. This raises the question of how different concentrations of morphogen could result in differences of transcriptional effectors activation through the same activation cascade. In numerous cases, the input-output response of morphogen-induced signaling pathway revealed a linear dependence between receptor activation level and transducer level in receiving cells. For example, in Xenopus cells, a threshold difference in the absolute number of activated receptor is sufficient to specify concentration-dependent responses to activin. Therefore the activin gradient is converted in cells as a gradient of transducer activation which is sufficient to mediate differential responses. From this observation we can understand that ones of the major parameters influencing cells responses to morphogen gradient are the parameters determining the gradient level itself. These parameters are **concentration at source $C0$, degradation rate $k$ and diffusion coefficient** $D$.

Different mechanism have been proposed and a major one is the **difference of binding affinity** of the transcription factor to different gene promotors. Indeed, low affinity binding site are occupied only at high concentration of transcription factor whereas high binding affinity site are bound even at low concentrations. 
As the signaling pathway is mostly linear the concentration of transcription factor is directly related to the morphogen concentration that is sensed by the cell. Consequently low morphogen concentration lead to a low concentration of specific transcription factors in cells. At low concentration transcription factor would only be able to activate gene with high binding affinity to this trancription factor while at high concentration it can activate both high and low binding affinity promotors.  

To illustrate this idea, we can look at the transcription activation of specific effectors in cells depending on the morphogen concentration. 
The probability of gene transcription activation $p$ by a transcription factor $X$ could be described by the Hill function. It depends on transcription factor concentration $[X]$, hill number $n$ and the reaction dissociation constant $K$ which correspond to the inverse of binding affinity.

\begin{align}
\ p([X]) = \frac{[X]^{n}}{[X]^{n} + K^{n}}
\end{align}
    
Here we consider a simplistic pathway where the interaction of a morphogen with its receptor results in the activation of a transcption factor $X$. This transcription factor control the transcription activation of two effectors, $A$ and $B$, which are responsible for cell identity selction. When $B$ is expressed at high level in cell it specifies $B$ cell identity while when $A$ and $B$ are co-expressed in cells it specifies A cell identity. $X$ binds to $A$ and $B$ genes with a binding affinity of $\frac{1}{K1}$ and $\frac{1}{K2}$, respectively. For the sake of simplicity we consider that morphogen concentration is directly converted in trancription factor concentration without taking into account intermediate steps in the pathway i.e. $C_{morphogen}(x)=[X](x)$. In this model activations thresholds are determined as the morphogen concentrations at which the probability of effector transcription activation reaches $p=0.9$ (i.e. high transcription).

### Morphogen interpretation depends on tissue-specific competence of cells

</div>

We saw that the cellular response to inductive event depends on the inducer level and its distribution in the tissue. However another major component of inductive event is the competence of the receiving tissue to interpret the signal. Such a competence to respond to inductive molecules relies on the gene expression of the cell. A cell can generate a specific outcome only if all the component to receive and transduce the signal are present within the cell. It means that the loss of competence occurs if some parts of th signal transduction cascade or the receptors are missing or blocked. These mechanisms explain the ON/OFF switch of signaling pathway activation in a tissue exposed to the same morphogen.

To illustrate this idea, we can take the same model of effector activation as before but this time we arbitrary set that after a distance limit $L_{comp.}$ from the source cells, receiving cells do not express the specific receptor of the morphogen. In this case the probability of effector activation becomes:

$$p([X]) =
  \begin{cases}
    \frac{[X]^{n}}{[X]^{n} + K^{n}}     & \quad \text{if x< } L_{comp.}\\
    0  & \quad \text{if x> } L_{comp.}
  \end{cases}
$$

### Influence of signaling dynamic

Morphogen gradients are not only regulated in space but also in time. Until now we only considered the morphogen gradient in its steady-state. However if the spread or clearance of the morphogen are realatively slow, it will take a period of time to reach this steady-state. During the pre-steady state the gradient concentration change at different rate at different positions in the tissue. Cells integrate its concentration over time resulting in an increase of the "sensed" morphogen concentration.

In addition, temporal scale generally also defined specific developmental stages due to temporally limited competence of the receiving cells. The duration of time during which cells can respond to a developmental signal is called "competency windows". These windows arise in response to pathway-intrinsic feedback which result in self-limiting competency.
